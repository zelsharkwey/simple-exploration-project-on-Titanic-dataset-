#### Q_5: Fill missing values in the Age column with the mean age then Group the data by Pclass and calculate the average age for each class

```python
m_mvalues = df['Age'].mean()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
average_age_class = df.groupby('Pclass')['Age'].mean()
average_age_class
```

```python
average_age_class.plot(kind='bar',color=['red','green','blue'])
plt.style.use('seaborn-v0_8')
plt.show()
```
